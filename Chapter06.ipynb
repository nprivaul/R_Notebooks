```r
lambda=2; Fm<- function(z) {1-exp(-lambda * z)}; max= 3; z <- seq(0, max, length.out= 1000)
plot(z,Fm(z),xlim= c(0,max),type='l',ylim= c(0,max),ylab="",main="", col= "blue",
lwd=2.8,xaxs="i",yaxs="i", cex.axis=1.6, cex.lab=1.6)
inverse <- function(z) {-log(1-z)/lambda}; lines(z,inverse(z), col= "purple", lwd=2.8)
lines(z,z, col= "red", lwd=2); abline(h=1,lwd=1); abline(v=1,lwd=1); grid()
```

```r
library(quantmod); getSymbols("0700.HK",from="2010-01-03",to="2018-02-01",src="yahoo")
stock=Ad(`0700.HK`);chartSeries(stock,up.col="blue",theme="white")
stk.rtn=(stock-lag(stock))/lag(stock)[-1]; returns <- as.numeric(stk.rtn[!is.na(stk.rtn)])
VaR <- function(p){r=-returns; ordered<-r[order(r)]; v=tail(ordered,1); i=length(r)
while(length(r[r<=v])/length(r)>=p) {i=i-1;VaR=v;v=ordered[i]}; return (-VaR)}
var1=VaR(0.95); var2=-quantile(-returns, probs = 0.95,type=1);times=index(stock)
length(returns[returns<var1])/length(returns); length(returns[returns<var2])/length(returns)
chartSeries(stk.rtn,up.col="blue",theme="white");abline(h=var1,col="red",lwd=3)
```
