```r
# install.packages("copula"); install.packages("gumbel")
library(copula);library(gumbel);norm.cop<-normalCopula(.35)
persp(norm.cop,pCopula,51,xlab="u",ylab="v",zlab="C(u,v)",col='mediumpurple1')
persp(norm.cop,dCopula,51,xlab="u",ylab="v",zlab="c(u,v)",col='mediumpurple1')
norm<-rCopula(4000,normalCopula(.7));plot(norm,cex=3,pch='.',col='blue')
stud<-rCopula(4000,tCopula(.5,df=1));points(stud,cex=3,pch='.',col='red')
gumb<-rgumbel(4000,4);points(gumb,cex=3,pch='.',col='green')
```

```r
joint_hist <- function(u) {x <- u[,1]; y <- u[,2]; xhist <- hist(x, breaks=40,plot=FALSE)
yhist <- hist(y, breaks=40,plot=FALSE); top <- max(c(xhist$counts, yhist$counts))
nf <- layout(matrix(c(2,0,1,3),2,2,byrow=TRUE), c(3,1), c(1,3), TRUE); par(mar=c(3,3,1,1))
plot(x, y, xlab="", ylab="",col="blue",pch=19,cex=.4)
points(x,-.01+rep(min(y),length(x)), xlab="", ylab="",pch=18,cex=.8)
points(-.01+rep(min(x),length(y)), y, xlab="", ylab="",pch=18,cex=.8); par(mar=c(0,3,1,1))
barplot(xhist$counts,axes=F,ylim=c(0,top),space=0,col="purple");par(mar=c(3,0,1,1))
barplot(yhist$counts, axes=F, xlim=c(0, top), space=0, horiz=TRUE,col="purple")}
joint_hist(norm);joint_hist(stud);joint_hist(gumb)
```

```r
set.seed(100);N=10000;m=c("norm","norm");p=list(list(mean=0,sd=1),list(mean=0,sd=1))
gaussM<-mvdc(normalCopula(.8),m,p);norm<-rMvdc(N,gaussM)
studentM<-mvdc(tCopula(.5,df=1),m,p);stud<-rMvdc(N,studentM)
gumbelM<-mvdc(gumbelCopula(4),m,p);gmb<-rMvdc(N,gumbelM)
plot(norm,pch='.',cex=3,col='blue');points(norm[,1],-.01+rep(min(norm[,2]),N),pch=18,cex=.8);
points(-.01+rep(min(norm[,1]),N),norm[,2],pch=18,cex=.8);joint_hist(norm)
plot(stud,pch='.',cex=3,col='blue');points(stud[,1],-.01+rep(min(stud[,2]),N),pch=18,cex=.8);
points(-.01+rep(min(stud[,1]),N),stud[,2],pch=18,cex=.8);joint_hist(stud)
plot(gmb,pch='.',cex=3,col='blue');points(gmb[,1],-.01+rep(min(gmb[,2]),N),pch=18,cex=.8);
points(-.01+rep(min(gmb[,1]),N),gmb[,2],pch=18,cex=.8);joint_hist(gmb)
```
