```r
library(quantmod); getSymbols("^HSI",from="2013-06-01",to="2014-10-01",src="yahoo")
data=Ad(`HSI`);returns=(data-lag(data))/lag(data);returns=as.numeric(returns[!is.na(returns)])
VaR(.95);var=-quantile(-returns,probs=.95,type=1);cte=mean(returns[returns<as.numeric(var)])
```

```r
mean(returns[returns<(VaR(0.95))],na.rm=TRUE)
```

```r
q=qnorm(.95, mean=0, sd=1); mu=mean(returns,na.rm=TRUE)
sigma=sd(returns,na.rm=TRUE); mu-sigma*dnorm(q)/0.05
```

```r
ES <- function(p){r=-returns; return (-(-VaR(p)+
(mean(r[r>=(-VaR(p))])-(-VaR(p)))*length(r[r>=(-VaR(p))])/length(r)/ (1-p)))}
ES(0.95)
```

```r
TVaR <- function(p){r=-returns; ordered=r[order(r)]; l=length(ordered); i=l; TVaR=0;
v=ordered[i];
while(length(r[r<=v])/length(r)>=p) {v=ordered[i]; j=1; while (v==ordered[i-j]) {j=j+1}; i=i-j;
TVaR=TVaR+v*min(j/l,(i+j)/l-p); v=ordered[i]}; return (-TVaR/(1-p))}
TVaR(0.95)
```
