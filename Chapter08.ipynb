```r
P <- function(S, K, mu, T, sigma)
{d1 <- (log(S/K)+(mu-sigma^2/2)*T)/(sigma*sqrt(T)); P= pnorm(-d1);P}
P(100,90,0.2,1,0.3)
```

```r
DBond <- function(S, K, r, T, sigma)
{d1= (log(S/K)+(r+sigma^2/2)*T)/(sigma*sqrt(T));d2= d1 - sigma * sqrt(T);
DBond= K*exp(-r*T) * pnorm(d2) - S*pnorm(-d1);DBond}
DBond(100, 105, 0.05, 1, 0.2)
```
